### Selección de Pozo y Sección para la Gráfica

Aquí puedes especificar el `Pozo` y la `Section` que deseas analizar. Basado en los datos disponibles, te sugiero usar `'FRUf19y'` como pozo y `12.25` como sección, pero puedes modificar estos valores. La gráfica mostrará `Bit depth (ft)` vs `Hours`, con las etiquetas del eje X indicando el `Day` y las líneas de cuadrícula para cada día.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# --- Configuración para seleccionar el pozo y la sección ---
selected_pozo = 'FRUf19y'  # @param {type:"string"}
selected_section = 12.25  # @param {type:"number"}

# --- Configuración para los límites de la gráfica ---
y_min_limit = 6800 # @param {type:"number"} Default: None (auto)
y_max_limit = 9200 # @param {type:"number"} Default: None (auto)
x_min_day = None # @param {type:"number"} Default: None (auto)
x_max_day = None # @param {type:"number"} Default: None (auto)

# --- Nuevos límites para la tasa de Cavings ---
cavings_min_limit = 0 # @param {type:"number"} Default: None (auto)
cavings_max_limit = 10 # @param {type:"number"} Default: None (auto)

# --- Nuevos límites para el sub-plot de cavings --- (No pedido, pero útil si se quiere)
cavings_size_min_limit = 0 # @param {type:"number"} Default: None (auto)
cavings_size_max_limit = 120 # @param {type:"number"} Default: None (auto)

# --- Nuevos límites para el sub-plot de MW/ECD ---
mw_ecd_min_limit = 8 # @param {type:"number"} Default: None (auto)
mw_ecd_max_limit = 13 # @param {type:"number"} Default: None (auto)

# Filtrar el DataFrame para el pozo y la sección seleccionados
df_filtered = df[(df['Pozo'] == selected_pozo) & (df['Section'] == selected_section)].copy()

# Convert None strings to actual None objects for parameter fields
if isinstance(x_min_day, str) and x_min_day.strip().lower() == 'none': x_min_day = None
if isinstance(x_max_day, str) and x_max_day.strip().lower() == 'none': x_max_day = None
if isinstance(cavings_min_limit, str) and cavings_min_limit.strip().lower() == 'none': cavings_min_limit = None
if isinstance(cavings_max_limit, str) and cavings_max_limit.strip().lower() == 'none': cavings_max_limit = None
if isinstance(cavings_size_min_limit, str) and cavings_size_min_limit.strip().lower() == 'none': cavings_size_min_limit = None
if isinstance(cavings_size_max_limit, str) and cavings_size_max_limit.strip().lower() == 'none': cavings_size_max_limit = None
if isinstance(mw_ecd_min_limit, str) and mw_ecd_min_limit.strip().lower() == 'none': mw_ecd_min_limit = None
if isinstance(mw_ecd_max_limit, str) and mw_ecd_max_limit.strip().lower() == 'none': mw_ecd_max_limit = None


if not df_filtered.empty:
    # Asegurar que 'Date' sea datetime y 'Day' sea numérico para ordenar
    df_filtered['Date'] = pd.to_datetime(df_filtered['Date'])
    df_filtered = df_filtered.sort_values(by=['Hours'])

    # Aplicar límites de días de operación si están definidos
    if x_min_day is not None: df_filtered = df_filtered[df_filtered['Day'] >= x_min_day]
    if x_max_day is not None: df_filtered = df_filtered[df_filtered['Day'] <= x_max_day]

    # Convertir las columnas 'MIN', 'MAX', 'MODE' a tipo numérico y rellenar NaN con 0
    for col in ['MIN', 'MAX', 'MODE']:
        if col in df_filtered.columns:
            df_filtered[col] = pd.to_numeric(df_filtered[col], errors='coerce').fillna(0)

    # --- NEW: Morphology columns numeric conversion and NaN handling ---
    # Reordered morphology_cols as per user request
    morphology_cols = ["RWK", "TAB", "BLOK", "ANG", "SPLIN", "SLK", "GRUM"]
    for col in morphology_cols:
        if col in df_filtered.columns:
            df_filtered[col] = pd.to_numeric(df_filtered[col], errors='coerce').fillna(0)
        else:
            print(f"Advertencia: La columna de morfología '{col}' no existe en los datos filtrados para el pozo '{selected_pozo}' y sección '{selected_section}'.")

    # Convert 'Cavings' and 'Cuttings+Cavings' to numeric, coercing errors and filling NaNs with 0 for safe calculation
    if 'Cavings' in df_filtered.columns:
        df_filtered['Cavings'] = pd.to_numeric(df_filtered['Cavings'], errors='coerce').fillna(0)
    else:
        print(f"Advertencia: La columna 'Cavings' no existe en los datos filtrados para el pozo '{selected_pozo}' y sección '{selected_section}'. Se asumirá 0 para los cálculos de proporción.\n")
        df_filtered['Cavings'] = 0

    cuttings_cavings_col_name = 'Cuttings+Cavings'
    if cuttings_cavings_col_name in df_filtered.columns:
        df_filtered[cuttings_cavings_col_name] = pd.to_numeric(df_filtered[cuttings_cavings_col_name], errors='coerce').fillna(0)
    else:
        print(f"Advertencia: La columna '{cuttings_cavings_col_name}' no existe en los datos filtrados para el pozo '{selected_pozo}' y sección '{selected_section}'. No se podrá calcular la proporción de cavings.\n")
        df_filtered[cuttings_cavings_col_name] = 0 # Assume 0 if missing to allow downstream code to run

    # Crear la figura y los ejes para los CUATRO subplots (arriba, medio, morfología, proporciones)
    fig, (ax1, ax3, ax5, ax6) = plt.subplots(nrows=4, figsize=(17, 11), sharex=True, gridspec_kw={'height_ratios': [8, 4, 3, 3]})

    # Plotear 'Cavings' como un área rellena (sombra) en azul en el eje principal (ax1)
    ax1.fill_between(df_filtered['Hours'], 0, df_filtered['Cavings'], color='blue', alpha=0.3, label='Tasa de Cavings')
    ax1.plot(df_filtered['Hours'], df_filtered['Cavings'], color='black', linestyle='-', linewidth=1, label='_nolegend_')

    # Aplicar límites para la tasa de Cavings si están definidos (ahora en ax1)
    if cavings_min_limit is not None or cavings_max_limit is not None:
        ax1.set_ylim(cavings_min_limit, cavings_max_limit)
    else:
        ax1.set_ylim(bottom=0) # Asegurar que la sombra empiece en 0 si no hay límites definidos

    ax1.set_ylabel('Tasa de Cavings', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')
    ax1.grid(True, linestyle=':', alpha=0.7)

    # Crear un segundo eje Y para 'Bit depth (ft)' (derecha) en el subplot superior (ax1)
    ax2 = ax1.twinx()
    sns.lineplot(x='Hours', y='Bit depth (ft)', data=df_filtered, color='red', linestyle='-', marker='', ax=ax2, label='Profundidad de Broca (ft)', legend=False)

    # Invertir el eje Y secundario (profundidad)
    ax2.invert_yaxis()

    # Aplicar límites del eje Y secundario si están definidos (ahora en ax2)
    if y_min_limit is not None or y_max_limit is not None:
        ax2.set_ylim(y_max_limit, y_min_limit) # Invertir límites para eje y invertido

    ax2.set_ylabel('Profundidad de Broca (ft)', color='red')
    ax2.tick_params(axis='y', labelcolor='red')

    # Calcular las posiciones de las etiquetas del eje X para cada día
    # Tomamos el primer valor de 'Hours' para cada 'Day'
    day_ticks = df_filtered.groupby('Day')['Hours'].min().reset_index()
    ax1.set_xticks(day_ticks['Hours'])
    ax1.set_xticklabels(day_ticks['Day'].astype(str), rotation=45, ha='right')
    ax1.set_xlabel('') # Eliminar etiqueta del eje X del subplot superior

    # Configurar líneas de cuadrícula para cada día en el subplot superior
    for hour_val in day_ticks['Hours']:\
        ax1.axvline(x=hour_val, color='gray', linestyle='--', linewidth=0.7, alpha=0.6)

    # Calcular la posición X para las etiquetas de Topes y BS
    x_label_pos = df_filtered['Hours'].max() + (df_filtered['Hours'].max() - df_filtered['Hours'].min()) * 0.01 # Un pequeño offset a la derecha

    # --- Añadir Topes --- al subplot superior
    if 'Pozo' in df_topes_dips.columns and 'MD' in df_topes_dips.columns:
        topes_pozo = df_topes_dips[df_topes_dips['Pozo'] == selected_pozo].copy()
        if y_min_limit is not None:
            topes_pozo = topes_pozo[topes_pozo['MD'] >= y_min_limit]
        if y_max_limit is not None:
            topes_pozo = topes_pozo[topes_pozo['MD'] <= y_max_limit]

        for _, row in topes_pozo.iterrows():
            ax2.axhline(y=row['MD'], color='black', linestyle='--', linewidth=1)
            ax2.text(x_label_pos, row['MD'], f"{row['Tope']}", color='black', verticalalignment='center', horizontalalignment='left', fontsize=9)

    # --- Añadir BS (Bottoms) --- al subplot superior
    if 'Pozo' in df_bs.columns and 'MD_sect' in df_bs.columns:
        bs_pozo = df_bs[df_bs['Pozo'] == selected_pozo].copy()
        if y_min_limit is not None:
            bs_pozo = bs_pozo[bs_pozo['MD_sect'] >= y_min_limit]
        if y_max_limit is not None:
            # Corrected column name from 'MD' to 'MD_sect'
            bs_pozo = bs_pozo[bs_pozo['MD_sect'] <= y_max_limit]

        for _, row in bs_pozo.iterrows():
            ax2.axhline(y=row['MD_sect'], color='blue', linestyle='--', linewidth=1)
            ax2.text(x_label_pos, row['MD_sect'], f"BS: {row['Sección']}\n", color='blue', verticalalignment='center', horizontalalignment='left', fontsize=9)

    ax1.set_title(f'Profundidad de Broca y Tasa de Cavings vs Horas para Pozo: {selected_pozo}, Sección: {selected_section}')

    # --- Configuración del Sub-plot del medio (ax3) ---

    # Plot MW and ECD on ax3 (left y-axis)
    if 'MW' in df_filtered.columns:
        if df_filtered['MW'].count() > 0: # Check if there's any non-NaN data
            ax3.plot(df_filtered['Hours'], df_filtered['MW'], color='red', linestyle='-', label='MW', linewidth=4, zorder=100) # Vivid red and thicker

            # Add labels for MW values where they change or for long constant segments
            mw_values = df_filtered['MW'].values
            mw_hours = df_filtered['Hours'].values

            # Find indices where MW values change
            change_indices = np.where(mw_values[:-1] != mw_values[1:])[0] + 1

            # Define segment start and end points (inclusive of start, exclusive of end for slicing)
            segment_start_indices = np.insert(change_indices, 0, 0) # Add 0 as the first start
            segment_end_indices = np.append(change_indices, len(mw_values)) # Add end of array as the last end

            min_segment_length = 3 # Minimum number of data points for a segment to get a label

            # Place text labels
            for seg_start, seg_end in zip(segment_start_indices, segment_end_indices):
                segment_length = seg_end - seg_start
                if segment_length >= min_segment_length:
                    # Place label in the middle of the segment
                    mid_idx = seg_start + segment_length // 2
                    x_pos = mw_hours[mid_idx]
                    y_pos = mw_values[mid_idx] # MW value is constant across the segment

                    ax3.text(x_pos, y_pos + 0.3,  # Increased offset for visual clarity
                             f'{y_pos:.1f}',       # Format MW value
                             color='black', fontsize=10, ha='center', va='bottom', fontweight='bold', zorder=10000) # Changed fontsize to 10

            # If there's only one constant MW value across the entire filtered data, ensure it's labeled
            if len(segment_start_indices) == 1 and segment_start_indices[0] == 0 and segment_end_indices[0] == len(mw_values) and len(mw_values) > 0:
                x_pos = mw_hours[len(mw_values) // 2]
                y_pos = mw_values[0]
                ax3.text(x_pos, y_pos + 0.3,  # Increased offset for visual clarity
                         f'{y_pos:.1f}',       # Format MW value
                         color='black', fontsize=10, ha='center', va='bottom', fontweight='bold', zorder=10000) # Changed fontsize to 10

        else:
            print(f"Advertencia: No hay datos válidos para 'MW' en el pozo '{selected_pozo}' y sección '{selected_section}'. No se mostrará en la gráfica.\n")
    else:
        print(f"Advertencia: La columna 'MW' no existe en los datos filtrados para el pozo '{selected_pozo}' y sección '{selected_section}'.\n")

    if 'ECD' in df_filtered.columns:
        if df_filtered['ECD'].count() > 0: # Check if there's any non-NaN data
            ax3.plot(df_filtered['Hours'], df_filtered['ECD'], color='blue', linestyle='-', label='ECD', linewidth=2, zorder=99) # Changed zorder to 99
        else:
            print(f"Advertencia: No hay datos válidos para 'ECD' en el pozo '{selected_pozo}' y sección '{selected_section}'. No se mostrará en la gráfica.\n")
    else:
        print(f"Advertencia: La columna 'ECD' no existe en los datos filtrados para el pozo '{selected_pozo}' y sección '{selected_section}'.\n")

    ax3.set_ylabel('MW / ECD', color='green')
    ax3.tick_params(axis='y', labelcolor='green')

    # Apply limits for MW/ECD on ax3
    if mw_ecd_min_limit is not None or mw_ecd_max_limit is not None:
        ax3.set_ylim(mw_ecd_min_limit, mw_ecd_max_limit)

    # Create a secondary y-axis for Cavings sizes on the right
    ax4 = ax3.twinx()

    # Plot MAX, MODE, MIN on ax4 (right y-axis)
    if 'MAX' in df_filtered.columns: # Asegurarse de que las columnas existan
        ax4.fill_between(df_filtered['Hours'], 0, df_filtered['MAX'], color='darkred', alpha=0.7, label='MAX Cavings Size', edgecolor='black', linewidth=0.5, zorder=2)
    if 'MODE' in df_filtered.columns: # Asegurarse de que las columnas existan
        ax4.fill_between(df_filtered['Hours'], 0, df_filtered['MODE'], color='yellow', alpha=0.7, label='MODE Cavings Size', edgecolor='black', linewidth=0.5, zorder=2)
    if 'MIN' in df_filtered.columns: # Asegurarse de que las columnas existan
        ax4.fill_between(df_filtered['Hours'], 0, df_filtered['MIN'], color='blue', alpha=0.7, label='MIN Cavings Size', edgecolor='black', linewidth=0.5, zorder=2)

    ax4.set_ylabel('Tamaño de Cavings', color='blue') # Set label for the right y-axis
    ax4.tick_params(axis='y', labelcolor='blue')

    # Apply limits for the Cavings sizes on ax4
    if cavings_size_min_limit is not None or cavings_size_max_limit is not None:
        ax4.set_ylim(cavings_size_min_limit, cavings_size_max_limit)
    else:
        ax4.set_ylim(bottom=0)

    # Configurar líneas de cuadrícula para cada día en el subplot medio
    ax3.set_xticks(day_ticks['Hours'])
    ax3.set_xticklabels(day_ticks['Day'].astype(str), rotation=45, ha='right')
    for hour_val in day_ticks['Hours']:
        ax3.axvline(x=hour_val, color='gray', linestyle='--', linewidth=0.7, alpha=0.6)
    ax3.grid(True, linestyle=':', alpha=0.7)
    ax3.set_xlabel('') # Eliminar etiqueta del eje X del subplot medio, ya que ax6 la tendrá

    # --- NEW: Configuración del Sub-plot inferior para Morfologías (ax5) ---
    morphology_colors = {
        "RWK": "red",
        "TAB": "darkgreen",
        "BLOK": "lime", # Fluorescent green
        "SPLIN": "blue",
        "ANG": "yellow",
        "SLK": "purple",
        "GRUM": "brown"
    }

    y_stack_bottom = np.zeros(len(df_filtered))

    for col in morphology_cols:
        if col in df_filtered.columns:
            # Plot the current morphology on top of the previous ones
            ax5.fill_between(df_filtered['Hours'], y_stack_bottom, y_stack_bottom + df_filtered[col],
                             color=morphology_colors.get(col, 'gray'), # Default to gray if color not specified
                             alpha=0.8, label=f'Morfología {col}', edgecolor='none') # Added edgecolor='none'
            # Update the base for the next morphology
            y_stack_bottom += df_filtered[col]

    ax5.set_ylabel('Porcentaje de Morfología (%)')
    ax5.grid(True, linestyle=':', alpha=0.7)
    ax5.set_ylim(0, 100) # Percentages should go from 0 to 100
    ax5.set_title('Distribución de Morfologías de Cavings a lo largo del tiempo')

    # X-axis ticks and labels for ax5, similar to ax1 and ax3
    ax5.set_xticks(day_ticks['Hours'])
    ax5.set_xticklabels(day_ticks['Day'].astype(str), rotation=45, ha='right')
    for hour_val in day_ticks['Hours']:\
        ax5.axvline(x=hour_val, color='gray', linestyle='--', linewidth=0.7, alpha=0.6)
    ax5.set_xlabel('') # Eliminar etiqueta del eje X del subplot de morfología, ya que ax6 la tendrá

    # --- NEW: Subplot para Proporción de Cavings vs Cuttings (ax6) ---
    df_filtered['Cuttings_Cavings_Total'] = df_filtered[cuttings_cavings_col_name]

    # Initialize proportions with NaN so fill_between doesn't plot where data is missing or total is zero
    df_filtered['Cavings_Proportion'] = np.nan
    df_filtered['Remainder_Proportion'] = np.nan

    # Calculate proportions only where Cuttings_Cavings_Total is greater than 0
    valid_indices = (df_filtered['Cuttings_Cavings_Total'] > 0) & (~df_filtered['Cavings'].isna())

    if valid_indices.any():
        df_filtered.loc[valid_indices, 'Cavings_Proportion'] = (\
            df_filtered.loc[valid_indices, 'Cavings'] / \
            df_filtered.loc[valid_indices, 'Cuttings_Cavings_Total']
        ) * 100
        df_filtered.loc[valid_indices, 'Remainder_Proportion'] = 100 - df_filtered.loc[valid_indices, 'Cavings_Proportion']

    # Plot Cavings Proportion
    ax6.fill_between(df_filtered['Hours'], 0, df_filtered['Cavings_Proportion'],
                     color='blue', alpha=0.8, label='Proporción Cavings')
    # Plot Remainder Proportion
    ax6.fill_between(df_filtered['Hours'], df_filtered['Cavings_Proportion'], 100,
                     color='yellow', alpha=0.8, label='Proporción Cuttings (Restante)')

    ax6.set_ylabel('Proporción (%)')
    ax6.set_xlabel('Día de Operación (Hours)') # Etiqueta X solo en el subplot inferior
    ax6.grid(True, linestyle=':', alpha=0.7)
    ax6.set_ylim(0, 100)
    ax6.set_title('Proporción de Cavings y Cuttings a lo largo del tiempo')

    # X-axis ticks and labels for ax6
    ax6.set_xticks(day_ticks['Hours'])
    ax6.set_xticklabels(day_ticks['Day'].astype(str), rotation=45, ha='right')
    for hour_val in day_ticks['Hours']:
        ax6.axvline(x=hour_val, color='gray', linestyle='--', linewidth=0.7, alpha=0.6)

    # --- Modificaciones para las leyendas de cada subplot ---

    # Leyenda para el subplot superior (ax1 / ax2)
    handles_ax1, labels_ax1 = ax1.get_legend_handles_labels()
    handles_ax2, labels_ax2 = ax2.get_legend_handles_labels()

    combined_handles_top = []
    combined_labels_top = []
    for h, l in zip(handles_ax1, labels_ax1):
        if l != '_nolegend_': # Excluir la etiqueta interna para 'Cavings' si es '_nolegend_'
            combined_handles_top.append(h)
            combined_labels_top.append(l)
    combined_handles_top.extend(handles_ax2)
    combined_labels_top.extend(labels_ax2)

    if combined_handles_top:
        ax1.legend(combined_handles_top, combined_labels_top, loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=len(combined_labels_top), fancybox=True, shadow=True, fontsize=10)

    # Leyenda para el subplot medio (ax3 / ax4)
    handles_ax3, labels_ax3 = ax3.get_legend_handles_labels()
    handles_ax4, labels_ax4 = ax4.get_legend_handles_labels()

    combined_handles_middle = handles_ax3 + handles_ax4
    combined_labels_middle = labels_ax3 + labels_ax4

    if combined_handles_middle:
        ax3.legend(combined_handles_middle, combined_labels_middle, loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=len(combined_labels_middle), fancybox=True, shadow=True, fontsize=10)

    # Leyenda para el subplot de morfología (ax5)
    handles_ax5, labels_ax5 = ax5.get_legend_handles_labels()
    labels_ax5_clean = [label.replace('Morfología ', '') for label in labels_ax5] # Eliminar "Morfología "

    if handles_ax5:
        ax5.legend(handles_ax5, labels_ax5_clean, loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=len(labels_ax5_clean), fancybox=True, shadow=True, fontsize=10)

    # Leyenda para el nuevo subplot de proporción (ax6)
    handles_ax6, labels_ax6 = ax6.get_legend_handles_labels()
    if handles_ax6:
        ax6.legend(handles_ax6, labels_ax6, loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=len(labels_ax6), fancybox=True, shadow=True, fontsize=10)


    # Ajustar layout para dejar espacio para las leyendas de cada subplot
    fig.tight_layout()
    plt.show()
else:
    print(f"No se encontraron datos para el Pozo: {selected_pozo} y Sección: {selected_section} con los límites especificados.")
```
